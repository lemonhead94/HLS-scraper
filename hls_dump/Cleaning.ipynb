## Actual Cleaning of the birth date

```python
df_scrape["cleaned_birth_date"] = df_scrape.birth_date.str.extract(r"((\d{1,2}\.\d{1,2}\.\d{3,4})|(\d{3,4}))")[0]
df_scrape.loc[11695, "cleaned_birth_date"] = "1.1.1375"
df_scrape.loc[13809, "cleaned_birth_date"] = "15.6.1833"
df_scrape.loc[14570, "cleaned_birth_date"] = "1.1.850"
df_scrape.loc[16168, "cleaned_birth_date"] = "1.1.1475"
df_scrape.loc[18142, "cleaned_birth_date"] = "1.1.400"
```

```python
df_scrape["cleaned_birth_date"]
```

# modify cleaned birth date to be in the format YYYY-MM-DD (format="%Y-%m-%d")

```python
# 8.8.1599 -> 1599-08-08
df_scrape["cleaned_birth_date_formatted"] = df_scrape[df_scrape["cleaned_birth_date"].str.fullmatch(r"(\d{1}\.\d{1}\.\d{4})", na=False)]["cleaned_birth_date"].str.replace(r"(\d{1})\.(\d{1})\.(\d{4})", r"\3-0\2-0\1", regex=True)
# 8.8.900 -> 0900-08-08
df_scrape["cleaned_birth_date_formatted"].fillna(df_scrape[df_scrape["cleaned_birth_date"].str.fullmatch(r"(\d{1}\.\d{1}\.\d{3})", na=False)]["cleaned_birth_date"].str.replace(r"(\d{1})\.(\d{1})\.(\d{3})", r"0\3-0\2-0\1", regex=True), inplace=True)
# 14.5.1900 -> 1900-05-14
df_scrape["cleaned_birth_date_formatted"].fillna(df_scrape[df_scrape["cleaned_birth_date"].str.fullmatch(r"(\d{2}\.\d{1}\.\d{4})", na=False)]["cleaned_birth_date"].str.replace(r"(\d{2})\.(\d{1})\.(\d{4})", r"\3-0\2-\1", regex=True), inplace=True)
# 14.5.900 -> 0900-05-14
df_scrape["cleaned_birth_date_formatted"].fillna(df_scrape[df_scrape["cleaned_birth_date"].str.fullmatch(r"(\d{2}\.\d{1}\.\d{3})", na=False)]["cleaned_birth_date"].str.replace(r"(\d{2})\.(\d{1})\.(\d{3})", r"0\3-0\2-\1", regex=True), inplace=True)
# 4.05.1900 -> 1900-05-04
df_scrape["cleaned_birth_date_formatted"].fillna(df_scrape[df_scrape["cleaned_birth_date"].str.fullmatch(r"(\d{1}\.\d{2}\.\d{4})", na=False)]["cleaned_birth_date"].str.replace(r"(\d{1})\.(\d{2})\.(\d{4})", r"\3-\2-0\1", regex=True), inplace=True)
# 4.05.500 -> 0500-05-04
df_scrape["cleaned_birth_date_formatted"].fillna(df_scrape[df_scrape["cleaned_birth_date"].str.fullmatch(r"(\d{1}\.\d{2}\.\d{3})", na=False)]["cleaned_birth_date"].str.replace(r"(\d{1})\.(\d{2})\.(\d{3})", r"0\3-\2-0\1", regex=True), inplace=True)
# 18.08.1500 -> 1500-08-18
df_scrape["cleaned_birth_date_formatted"].fillna(df_scrape[df_scrape["cleaned_birth_date"].str.fullmatch(r"(\d{2}\.\d{2}\.\d{4})", na=False)]["cleaned_birth_date"].str.replace(r"(\d{2})\.(\d{2})\.(\d{4})", r"\3-\2-\1", regex=True), inplace=True)
# 18.08.900 -> 0900-08-18
df_scrape["cleaned_birth_date_formatted"].fillna(df_scrape[df_scrape["cleaned_birth_date"].str.fullmatch(r"(\d{2}\.\d{2}\.\d{3})", na=False)]["cleaned_birth_date"].str.replace(r"(\d{2})\.(\d{2})\.(\d{3})", r"0\3-\2-\1", regex=True), inplace=True)
# 900 to 0900-01-01
df_scrape["cleaned_birth_date_formatted"].fillna(df_scrape[df_scrape["cleaned_birth_date"].str.fullmatch(r"(\d{3})", na=False)]["cleaned_birth_date"].str.replace(r"(\d{3})", r"0\1-01-01", regex=True), inplace=True)
# 1500 to 1500-01-01
df_scrape["cleaned_birth_date_formatted"].fillna(df_scrape[df_scrape["cleaned_birth_date"].str.fullmatch(r"(\d{4})", na=False)]["cleaned_birth_date"].str.replace(r"(\d{4})", r"\1-01-01", regex=True), inplace=True)
```

```python
df_scrape["cleaned_death_date"] = df_scrape.death_date.str.extract(r"((\d{1,2}\.\d{1,2}\.\d{3,4})|(\d{3,4}))")[0]
# 8.8.1599 -> 1599-08-08
df_scrape["cleaned_death_date_formatted"] = df_scrape[df_scrape["cleaned_death_date"].str.fullmatch(r"(\d{1}\.\d{1}\.\d{4})", na=False)]["cleaned_death_date"].str.replace(r"(\d{1})\.(\d{1})\.(\d{4})", r"\3-0\2-0\1", regex=True)
# 8.8.900 -> 0900-08-08
df_scrape["cleaned_death_date_formatted"].fillna(df_scrape[df_scrape["cleaned_death_date"].str.fullmatch(r"(\d{1}\.\d{1}\.\d{3})", na=False)]["cleaned_death_date"].str.replace(r"(\d{1})\.(\d{1})\.(\d{3})", r"0\3-0\2-0\1", regex=True), inplace=True)
# 14.5.1900 -> 1900-05-14
df_scrape["cleaned_death_date_formatted"].fillna(df_scrape[df_scrape["cleaned_death_date"].str.fullmatch(r"(\d{2}\.\d{1}\.\d{4})", na=False)]["cleaned_death_date"].str.replace(r"(\d{2})\.(\d{1})\.(\d{4})", r"\3-0\2-\1", regex=True), inplace=True)
# 14.5.900 -> 0900-05-14
df_scrape["cleaned_death_date_formatted"].fillna(df_scrape[df_scrape["cleaned_death_date"].str.fullmatch(r"(\d{2}\.\d{1}\.\d{3})", na=False)]["cleaned_death_date"].str.replace(r"(\d{2})\.(\d{1})\.(\d{3})", r"0\3-0\2-\1", regex=True), inplace=True)
# 4.05.1900 -> 1900-05-04
df_scrape["cleaned_death_date_formatted"].fillna(df_scrape[df_scrape["cleaned_death_date"].str.fullmatch(r"(\d{1}\.\d{2}\.\d{4})", na=False)]["cleaned_death_date"].str.replace(r"(\d{1})\.(\d{2})\.(\d{4})", r"\3-\2-0\1", regex=True), inplace=True)
# 4.05.500 -> 0500-05-04
df_scrape["cleaned_death_date_formatted"].fillna(df_scrape[df_scrape["cleaned_death_date"].str.fullmatch(r"(\d{1}\.\d{2}\.\d{3})", na=False)]["cleaned_death_date"].str.replace(r"(\d{1})\.(\d{2})\.(\d{3})", r"0\3-\2-0\1", regex=True), inplace=True)
# 18.08.1500 -> 1500-08-18
df_scrape["cleaned_death_date_formatted"].fillna(df_scrape[df_scrape["cleaned_death_date"].str.fullmatch(r"(\d{2}\.\d{2}\.\d{4})", na=False)]["cleaned_death_date"].str.replace(r"(\d{2})\.(\d{2})\.(\d{4})", r"\3-\2-\1", regex=True), inplace=True)
# 18.08.900 -> 0900-08-18
df_scrape["cleaned_death_date_formatted"].fillna(df_scrape[df_scrape["cleaned_death_date"].str.fullmatch(r"(\d{2}\.\d{2}\.\d{3})", na=False)]["cleaned_death_date"].str.replace(r"(\d{2})\.(\d{2})\.(\d{3})", r"0\3-\2-\1", regex=True), inplace=True)
# 900 to 0900-01-01
df_scrape["cleaned_death_date_formatted"].fillna(df_scrape[df_scrape["cleaned_death_date"].str.fullmatch(r"(\d{3})", na=False)]["cleaned_death_date"].str.replace(r"(\d{3})", r"0\1-01-01", regex=True), inplace=True)
# 1500 to 1500-01-01
df_scrape["cleaned_death_date_formatted"].fillna(df_scrape[df_scrape["cleaned_death_date"].str.fullmatch(r"(\d{4})", na=False)]["cleaned_death_date"].str.replace(r"(\d{4})", r"\1-01-01", regex=True), inplace=True)

# fill missing dates with hhd_death_date
df_scrape.fillna({"cleaned_death_date_formatted": df_scrape.hhb_death_date}, inplace=True)
```
